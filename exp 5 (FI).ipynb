```
(* Task #1 玻尔兹曼近似 *)
f[η_] := Sqrt[η] Exp[-η]
Plot[f[η],{η,0,10}]
```

```
Integrate[f[η],{η,0,Infinity}]
%//N
```

```
(* 将二者进行比较，说出其本质区别。 *)
NIntegrate[f[η],{η,0,Infinity}]
```

```
(* Task #2 费米积分，玻尔兹曼近似条件失效 *)

f[η_] := η^(1/2) /(1 + Exp[η - ηF])
```

```
ηF = 1;
Plot[f[η],{η,0,Infinity}]
```

```
trapz[a_, b_, n_] := Module[{h, k, x},
    h = (b - a)/n;
    x[k_] := a + k h;
    N[
        h*(f[a] + f[b]) /2 + h Sum[f[x[k]],{k,1,n-1}]]]
```

```
(* 尝试着使用最基础的“梯形法则”算法解决。 *)
trapz[0,10,8000]
```

```
(* 设想如果仅该算法，那么怎么判断何时该终止投入算力。 *)
Do[trapz[0,10,i*100]//Print,{i,1,10,1}]
```

```
(* Task #3 制作一个用于计算费米积分的“函数”； *)

(* ηF = -6 ~ 10
    n = 0, 1/2, 1, 2 *)
    
FI[n_, ηF_] := NIntegrate[η^n /(1 + Exp[η - ηF]),{η,0,Infinity}]
```
