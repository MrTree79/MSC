```
U[r_] := A/r^12 - B/r^6
```

```
epsilon = 0.997;
sigma = 3.4;
A = 4 epsilon sigma^12;
B = 4 epsilon sigma^6;

Plot[U[r], {r,3,8}]
```

```
(* 方案一：使用 Solve *)
root = Solve[
    {D[U[r],r] == 0, r > 0}, r];  
    (* 使用Solve求解多项式方程，符号计算范畴；然而容易引发警告； *)

root[[1]]
U[r]/.root[[1]]
```

```
(* 方案二：使用 NSolve *)

root = NSolve[D[U[r], r] == 0, r]  (*使用NSolve代替Solve*)
root[[1]]
U[r]/.root[[1]]
```
