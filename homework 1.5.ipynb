```
f[x_] := ArcTan[x]//Sqrt
```

```
Plot[f[x],{x,0,1}, Filling -> Axis]
```

```
trapz[a_, b_, n_] := Module[{h, k, x},
    h = (b - a)/n;
    x[k_] := a + k h;
    N[
        h*(f[a]+f[b])/2 + h Sum[f[x[k]],{k,1,n-1}]
    ]
]
```

```
trapz[0,1,10]
```

```
NIntegrate[f[x],{x,0,1}]
```

```
showTrapz[xmin_,xmax_,n_]:=Block[{ps=Range[xmin,xmax,(xmax - xmin)/n],fI},
    ps = Table[{ps[[i]],ps[[i]]//f},{i,Length[ps]}];
    fI = Interpolation[ps, InterpolationOrder-> 1];
    
    Show[ListPlot[ps,
    	PlotStyle -> {Red,PointSize[0.02]},
    	Filling -> Axis],

     Plot[{f[x],fI[x]},{x,xmin,xmax},Filling-> Axis]]]
```

```
showTrapz[0,1.0,10]
```
